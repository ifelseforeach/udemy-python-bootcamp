## Matplotlib concepts

Very similar to MatLab graphic plotting

```python
import matplotlib.pyplot as plt
```

%matplotlib inline allows you to see plots inside the notebook, you don't need to use <b>plt.show()</b>

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

### 1) Functional method

```python
plt.plot(x,y, 'r-')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Plot title")
```

plt commands should be in one cell, overwise they will run separately

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

### 2) Object oriented method

add_axes agrs: left, bottom, right?,up?

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

### Several plots

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title('Large plot')

axes2=fig.add_axes([0.2,0.5,0.4,0.4])
axes2.plot(y,x)
axes2.set_title('Small plot')
```

### Subplots

```python
fig,axes = plt.subplots(1,2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
type(axes)
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(7,3))
axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y)
```

```python
fig, axes= plt.subplots(nrows=1, ncols=2,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

### Save a figure as a file (.jpg, .png, ...)

```python
fig.savefig('Figure.jpeg', dpi=300)
```

```python
fig= plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,x**2,label='Square')
axes.plot(x,x**3,label='Cube')

axes.legend(loc='best')
#axes.legend(loc=0)--location code
#axes.legend(loc=(0.1,0.1))-- tuple
```

### Setting colors

```python
fig=plt.figure()
ax=fig.add_axes([0.0,0.0,1.0,1.0])
#ax.plot(x,y,color='r')
#ax.plot(x,y,color='red')
#ax.plot(x,y,color='#FF8833',linewidth=5,alpha=0.5)
#ax.plot(x,y,color='purple',lw=3,linestyle='steps')
#linestyle: [ '-' | '--' | '-.' | ':' | 'steps' | ...]
ax.plot(x,y,color='blue',lw=3,ls='-',marker='*', ms=30, mfc='r',mec='g')
#marketsize = ms
#markerfacecolor or mfc
#marker: [ '+' | ',' | '.' | '1' | '2' | '3' | '4' | '*' | 'o' ]
```

### Limit region

```python
fig=plt.figure()
ax=fig.add_axes([0.0,0.0,1.0,1.0])
ax.plot(x,y,color='purple',lw=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
